```python
import pandas as pd
import numpy as np

df = pd.read_csv('YRBS_2007.csv')

# =========================================================================
# 1: Data Preparation & Recoding for the Dependent Variable (Y)
# =========================================================================
# The original 'Sleep' column contains questionnaire codes from 1 to 7.
# We map these categorical codes back to their actual continuous sleep hours

sleep_map = {
    1: 4,   # 4 hours or less
    2: 5,   # 5 hours
    3: 6,   # 6 hours
    4: 7,   # 7 hours
    5: 8,   # 8 hours
    6: 9,   # 9 hours
    7: 10   # 10 hours or more
}
df['Sleep_Hours'] = df['Sleep'].map(sleep_map)


# =========================================================================
# 2: Data Preparation & Recoding for the Independent Variable (X)
# =========================================================================
# The 'AerobicExercise' column ranges from 1 (0 days) to 8 (7 days).
# To make the comparison meaningful and avoid too many fragmented categories,
# we recode the 8 levels into 3 distinct exercise frequency groups.
def recode_aerobic(value):
    if value in [1, 2]:        # 0 to 1 day per week
        return 'Low'
    elif value in [3, 4, 5]:    # 2 to 4 days per week
        return 'Medium'
    elif value in [6, 7, 8]:    # 5 to 7 days per week
        return 'High'
    else:
        return np.nan           # Mark missing or invalid data as NaN

df['Exercise_Group'] = df['AerobicExercise'].apply(recode_aerobic)


# =========================================================================
# 3: Handle Missing Values & Filter Relevant Columns
# =========================================================================
df_final = df[['Exercise_Group', 'Sleep_Hours']].dropna()


# =========================================================================
# 4: Verify Cleaned Data Distribution
# =========================================================================
print("--- Distribution of Exercise Groups (X) ---")
print(df_final['Exercise_Group'].value_counts())

print("\n--- Descriptive Statistics for Sleep Hours (Y) ---")
print(df_final['Sleep_Hours'].describe())

df_final.to_csv('YRBS_2007_Cleaned.csv', index=False)
```
